```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline
from peft import PeftModel

num_labels = 7
id2label = {
    "0": "neutral",
    "1": "anger",
    "2": "disgust",
    "3": "fear",
    "4": "happiness",
    "5": "sadness",
    "6": "surprise"
}

label2id = {
    "neutral": 0,
    "anger": 1,
    "disgust": 2,
    "fear": 3,
    "happiness": 4,
    "sadness": 5,
    "surprise": 6
}

tokenizer = AutoTokenizer.from_pretrained("michellejieli/emotion_text_classifier")
base_model = AutoModelForSequenceClassification.from_pretrained("michellejieli/emotion_text_classifier",
                                                                num_labels=num_labels,
                                                                id2label=id2label,
                                                                label2id=label2id)
# base_model = AutoModelForSequenceClassification.from_pretrained("michellejieli/emotion_text_classifier", num_labels=num_labels, id2label=id2label, label2id=label2id)
# peft_model_id = "Shotaro30678/sentiment-analysis-ncu-chat-bot"
# model = PeftModel.from_pretrained(base_model, peft_model_id)
# model = model.merge_and_unload()
```
